### NUMERO DE IMAGENES 

```python
#Antes de todo queremos ver el numero medio de las variable economicas en funcion del cluter y del numero de imagenes, videos, formularios y enclaces
#para ver si influye de alguna forma

import pandas as pd
imagenes = datos['num_imagenes']
videos = datos['num_videos']
enlaces = datos['num_enlaces']
formularios = datos['num_formularios']

y1 = datos['media_numero_empleados']
y2 = datos['media_inmovilizado_intangible']
y3 = datos['media_rentabilidad_economica']
etiquetas = datos['cluster']

# Crear el DataFrame para almacenar los valores medios
df_media = pd.DataFrame(columns=['Variable', 'Cluster', 'ImÃ¡genes', 'Videos','Enlaces', 'Formularios'])

# Calcular el valor medio de imÃ¡genes por cluster para cada variable
for variable, datos_y in zip(['NÃºmero de empleados', 'Inmovilizado intangible', 'Rentabilidad econÃ³mica'], [y1, y2, y3]):
    for cluster in set(etiquetas):
        indices_cluster = etiquetas == cluster
        valor_medio_imagenes = imagenes[indices_cluster].mean()
        valor_medio_videos = videos[indices_cluster].mean()
        valor_medio_enlaces = enlaces[indices_cluster].mean()
        valor_medio_formularios = formularios[indices_cluster].mean()
        
        df_media = df_media.append({'Variable': variable, 'Cluster': cluster, 'ImÃ¡genes': valor_medio_imagenes,
                                   'Videos': valor_medio_videos, 'Enlaces': valor_medio_enlaces,
                                   'Formularios': valor_medio_formularios}, ignore_index=True)

print(df_media)
```

### QUE PALABRAS SE REPITEN MAS EN CADA CLUSTER

```python
def palabras_por_cluster(cluster, numero):
    lista_cluster = []
    for index,row in cluster.iterrows():
        lista_cluster.append(cluster['top10'][index])
        
    for lista in range (len(lista_cluster)):
        elemento = lista_cluster[lista]
        elemento = elemento.replace(' ', '')
        elemento = elemento.replace('[', '')
        elemento = elemento.replace(']', '')
        elemento = elemento.replace("'", '')
        lista_cluster[lista] = elemento
    
    #Cogemos todas las palabras
    todas_palabras = []
    for lista in range(len(lista_cluster)):
        for palabra in lista_cluster[lista].split(','):
            todas_palabras.append(palabra)

    from collections import Counter
    frecuencia = Counter(todas_palabras)
    top10 = frecuencia.most_common(10)

    print("Las 10 palabras mÃ¡s comunes del cluster", numero,  "son: " )
    for palabra, veces in top10:
        print(f"{palabra}: {veces} veces")
```

```python
palabras_por_cluster(cluster0, 0)
```

```python
palabras_por_cluster(cluster1, 1)
```

```python
palabras_por_cluster(cluster2, 2)
```

```python
palabras_por_cluster(cluster3, 3)
```

```python
palabras_por_cluster(cluster4, 4)
```
