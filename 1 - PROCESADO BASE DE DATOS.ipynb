#### Funciones para obtener el vector

```python
import requests
from bs4 import BeautifulSoup
import re
import openai
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Contador de las palabras que mas aparecen y las pone al inicio
from collections import Counter
import string
```

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("à", "a"),
        ("è", "e"),
        ("ì", "i"),
        ("ò", "o"),
        ("ù", "u"),)
    
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s
```

```python
def procesamiento_del_texto(url):
    #Download the HTML
    #allow_redirects : debe de estar en True para que el texto se imprima bien
    #Headers = para todas aquellas que devuelve una lista vacia pero si que accede a la web -> no afecta a las que ya funcionaban bien
    #verify = False : para acceder a todas esas webs que son privadas -> no afecta a las publicas
    
    #headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }
   
    res = requests.get('http://' + url, allow_redirects = True,  verify = False) #headers = headers
    html_page = res.content 
    
    #We use BeautifulSoup to parse the HTML
    soup = BeautifulSoup(html_page, 'html.parser')
    
    
    #Vamos a quedarnos con el cuerpo del texto 
    texto = []
    estructura_html = ['head', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6','p']
    for etiqueta in estructura_html:
        
        #encuentra todo el texto que pertenece a esa etiqueta
        p_tags = soup.find_all(etiqueta)
        for tag in p_tags:
            texto.append(tag.get_text().rstrip())
            
    #Quitamos los saltos de linea del texto
    texto_sin_salto_linea = []
    for line in texto:
        line = line.replace('\n', '')
        line = line.strip()
        texto_sin_salto_linea.append(line)
        
    #Separar minuscula -> mayuscula
    patron = r'([a-z])([A-Z])'
    texto_separado_min_may = []
    for lista in texto_sin_salto_linea:
        resultado = re.sub(patron, r'\1 \2', lista)
        texto_separado_min_may.append(resultado)
        
    #Vamos a quedarnos solo con las palabras -> fuera números y fuera carácteres especiales
    solo_palabras =[]
    for linea in texto_separado_min_may:
        linea = normalize(linea)
        linea = linea.lower()
        #elimina caracteres especiales
        linea = re.sub(r"\W"," ", linea)
        #elimina los numeros
        linea = re.sub("\d+", "", linea)
        linea = linea.strip()
        if linea != '':
            solo_palabras.append(linea)
            
    #Eliminar mas de un espacio en blanco
    texto_final = []
    for linea in solo_palabras:
        palabras = linea.split()
        frase = ' ' .join(palabras)
        texto_final.append(frase)
    #Ademas uno toda la cadena para tener todo el texto junto
    #Todo el texto procesado esta en esta variable -> texto_procesado_final
    texto_procesado_final = " ".join(texto_final) 
    
    
    
    
    #Obtenermos las variables extras que queremos -> IMAGENES, VIDEOS, ENLACES y FORMULARIOS
    contador_imagenes = []
    contador_video = []
    contador_enlace = []
    contador_formularios = [] #crea controles interactivos para formularios, con el fin de recibir datos del usuario
    
    for imagen in soup.find_all('img'):
        contador_imagenes.append(imagen.get('src'))
    for video in soup.find_all('video'):
        contador_video.append(video.get('src'))
    for enlace in soup.find_all('a'):
        contador_enlace.append(enlace.get('src'))
    for formulario in soup.find_all('input'):
        contador_formularios.append(formulario.get('src'))
        
    dic = {'Imagenes' : len(contador_imagenes), 
           'Videos' : len(contador_video), 
           'Enlaces' : len(contador_enlace),
           'Formularios' : len(contador_formularios)}
    
    
    
    
    #Obtenermos el vector de EMBEDDING
    openai.api_key = "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"
    if texto_procesado_final != '':
        vector = openai.Embedding.create(input = [texto_procesado_final], model = "text-embedding-ada-002")['data'][0]['embedding']
    else:
        vector = ''
    
    #Obtenemos las Top10 palabras mas importantes
    #Son las palabras que no interesan -> ej/ y, de, que
    stop_words_es = set(stopwords.words('spanish'))
    stop_words_en = set(stopwords.words('english'))
    #sacamos todas las palabras que hay en el texto
    word_tokens = word_tokenize(texto_procesado_final)
    word_tokens = list(filter(lambda token: token not in string.punctuation, word_tokens))
    #quitar las stop_words
    filtro = []
    for palabra in word_tokens:
        if palabra not in stop_words_es and palabra not in stop_words_en:
            filtro.append(palabra)
            
    c = Counter(filtro)
    diez_populares = c.most_common(10)
    
    lista_populares = []
    for palabra in diez_populares:
        lista_populares.append(palabra[0])
    
    
    
    #Creamos la lista final que queremos devolver -> ¿qué contiene?
    lista_variables_finales = []
    #Variables con informacion extra de la web
    lista_variables_finales.append(dic["Imagenes"])
    lista_variables_finales.append(dic["Videos"])
    lista_variables_finales.append(dic["Enlaces"])
    lista_variables_finales.append(dic["Formularios"])
    #Variable de vector de embedding
    lista_variables_finales.append(vector)
    #Variable de las palabras mas populares
    lista_variables_finales.append(lista_populares)
    
    return lista_variables_finales
```
